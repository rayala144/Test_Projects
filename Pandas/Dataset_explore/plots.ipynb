```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
sample_data = {'Year': [2010, 2011, 2012, 2013, 2014, 2015],
        'Sales': [2500, 2700, 2900, 3100, 3300, 3500]}

df = pd.DataFrame(sample_data)

# Generate the bar graph
df.plot(kind='bar', x='Year', y='Sales', color='green', figsize=(10, 6))

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales Data')

plt.show()

```

```python

# Sample dataset
data = {'Year': [2010, 2011, 2012, 2013, 2014, 2015],
        'Sales': [2500, 2700, 2900, 3100, 3300, 3500]}

df1 = pd.DataFrame(data, index=data['Year'])


# total_sales = df1['Sales'].sum()

# Generate the pie chart
df1['Sales'].plot(kind='pie', figsize=(5, 5), autopct=lambda x : '{:.0f}%'.format(x))

# Add labels and title to the graph
plt.axis('equal')
plt.legend(df1['Year'])
plt.title('Yearly Sales Data')

# Show the graph
plt.show()
```
